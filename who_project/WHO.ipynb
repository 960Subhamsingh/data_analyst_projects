Outliers  treat

```python
def sker(col):
    q1,q3 = np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw, uw
```

```python
sker(df['GDP'])
```

```python
for i in ['GDP','Total_expenditure',' thinness  1-19 years', ' thinness 5-9 years']:
    lw , uw = sker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>lw,lw,df[i])    
```
