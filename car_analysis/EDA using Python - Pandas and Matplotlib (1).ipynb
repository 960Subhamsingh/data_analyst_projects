```python
# Import the necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 
sns.set(color_codes=True)
```

```python
# Load the car.csv dataset
car = pd.read_csv('C:/Users/hp/Desktop/Data-Science-main/Data Analysis/Data Analysis using Python-20220118T163248Z-001/Data Analysis using Python/Data Analysis using Python/car.csv')
```

```python
# How many cars are available in each category of engine
car.groupby('Engine Fuel Type')['Engine Cylinders'].sum()
```

```python
car['No of Doors']=car['Number of Doors'].apply(lambda x:1 if(x<4.0) & (x>5.0) else 0)
```

```python
car['No of Doors'].interpolate()
```

```python
# Rows containing duplicate data
duplicate_rows_car = car[car.duplicated()]
print("Number of duplicate rows: ", duplicate_rows_car.shape)
```

```python
# Drop the duplicates 
car = car.drop_duplicates()
car.head()
```

```python
# Analytical summary of the dataset
car.describe(include='all')
```

```python
car.hist(figsize=(20,30))
```

```python
# Relationship between categorical and continuous variable
sns.boxplot(x="Vehicle Size", y="Engine HP", data = car)
```

```python
sns.pairplot(car)
```

```python
# Drop irrelevant columns
car = car.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
car.head(5)
```

```python
# Renaming the column names
car = car.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
car.head(5)
```
